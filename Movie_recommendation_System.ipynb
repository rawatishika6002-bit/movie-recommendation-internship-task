```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Movie dataset
movies = {
    "Inception": "sci-fi thriller mind-bending dream",
    "Interstellar": "sci-fi space time emotional drama",
    "The Dark Knight": "action crime superhero",
    "Avengers Endgame": "action superhero fantasy",
    "Titanic": "romance drama tragedy",
    "The Matrix": "sci-fi action virtual reality",
    "Joker": "psychological crime drama",

    "3 Idiots": "college friendship comedy drama motivation",
    "Dangal": "sports wrestling family motivation",
    "Kabir Singh": "romance emotional drama",
    "PK": "comedy drama social satire",
    "Zindagi Na Milegi Dobara": "friendship travel self discovery",
    "Bajrangi Bhaijaan": "emotional family drama",
    "War": "action spy thriller"
}

movie_names = list(movies.keys())
descriptions = list(movies.values())

# Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(descriptions)

# Similarity calculation
similarity = cosine_similarity(tfidf_matrix)

def recommend_movie(movie_name):
    if movie_name not in movie_names:
        print("‚ùå Movie not found in database.")
        return

    index = movie_names.index(movie_name)
    scores = list(enumerate(similarity[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nüé• Recommended movies for '{movie_name}':\n")
    for i in scores[1:6]:
        print("‚û°", movie_names[i[0]])

# User input
user_movie = input("Enter a movie name: ")
recommend_movie(user_movie)
```
